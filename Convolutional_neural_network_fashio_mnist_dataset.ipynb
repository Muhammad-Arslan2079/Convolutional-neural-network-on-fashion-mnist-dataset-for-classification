```python
# importing required libraries
import pandas as pd # pandas for preprocessing and data analysis
import torch # pytorch
from sklearn.model_selection import train_test_split # to split data into training and test data
import matplotlib.pyplot as plt # for visualizations
import torch.nn as nn # neural network module of pytorch
from torch.utils.data import Dataset,DataLoader # to load data and make batches of features and labels
import torch.optim as optim # optimizer fro parameter updates
import numpy as np # for numerical calculation
```

```python
# Cheking GPU availability ,step 1 for training on gpu
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f' using device:{device}')
```

```python
# setting random seed for reproducibility
torch.manual_seed(42)
```

```python
# load data
df = pd.read_csv('/content/fashion-mnist_train.csv')
df.head()
```

```python
# separating features from target labels
x= df.drop(columns='label',axis=1)
y=df['label']
```

```python
# spliting data into train,test
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
# scaling the features between 0 and 1 range of values for stability
x_train = x_train/255.0
x_test = x_test/255.0
```

```python
# our  tain,test split or dataframe type, pytorch expects tensors
# converting df type to numpy array and then numpy array to tensor
x_train = x_train.values
x_test = x_test.values
y_train = y_train.values
y_test = y_test.values
```

```python
print(type(x_train))
```

```python
# converting now np arrays to tensors
x_train = torch.from_numpy(x_train)
x_test = torch.from_numpy(x_test)
y_train = torch.from_numpy(y_train)
y_test = torch.from_numpy(y_test)
```

```python
x_train
```

```python
# creating custom dataset class
class CustomDataset(Dataset): # our custom class inherit base Dataset class
     def __init__(self,features,labels):
       self.features = torch.tensor(features,dtype=torch.float32).reshape(-1,1,28,28) #since we haive 784 columns/features,this dataset class will fetch row that is 1 dimensional,now we are working on images that are 2  dimensional, so reshaping as 28*28 makes 784 columns total,28 pixels left to right,28 pixels 28 pixels top to bottom, and 1 represent its 1 channel/gray scale image
       self.labels = torch.tensor(labels, dtype=torch.long) # attribute to store labels
      # pytorch expects features in float dtype and labels in long dtype means integers
     def __len__(self):
      return self.features.shape[0] # no. of rows
     def __getitem__(self,index):
      return self.features[index], self.labels[index] # return rows and  labels at provided index
```

```python
# creating train and test objects of custom class to load train and test data separatly from same custom class
train_dataset = CustomDataset(x_train,y_train) # arguments provided for features and labels parameters
test_dataset = CustomDataset(x_test,y_test)
```

```python
# cheking no. of rows by calling len method of custom class
len(train_dataset)
```

```python
# extracting rows by calling get item method of custom class
train_dataset[0]
```

```python
# creating train and test loader objects of DataLoader class
train_loader = DataLoader(train_dataset,batch_size= 128,shuffle= True,pin_memory= True)
test_loader = DataLoader(test_dataset,batch_size=128, shuffle= False, pin_memory= True) # we dont need to shuffle data in evaluation,beacuese it effect accuracies
```

```python
# defining neural network class
class Model(nn.Module): # our custom class inherit base nn module class to access its offerings
     def __init__(self,input_features): # we need no. of columns as input,to initialize code
       super().__init__()
       self.feature_extraction_layer = nn.Sequential(
      # first conv-pool layers pair
      nn.Conv2d(input_features,32,kernel_size=3,padding='same'), # input feature represents input channels as rgb/or grayscale,we provide values as 1 or 3,then output channels, means no. filters used in conv layer,expample 64 filters means 54 types of patterns, 1 filter detect 1 pattern,same padding means same size of output as = input, 0 padding so size after padding will be same, 3 means 3 by 3 kernal
      nn.ReLU(),
      nn.BatchNorm2d(32), # batch normalization on activation before passing to 2nd layer, 32 filters maps/output channels on which it is applied
      nn.MaxPool2d(kernel_size=2,stride=2), # kernal is 2 by 2 and stride/jump of 2
      #2nd conv-pool pair layer
      nn.Conv2d(32,64,kernel_size=3,padding='same'),# here input features/channel will be the output of previous layer, and output channels are 64 filters to be used to detect 64 types of patterns
      nn.ReLU(),
      nn.BatchNorm2d(64),
      nn.MaxPool2d(kernel_size=2,stride=2)

      # filters are feature maps from convolutional layers before passing to max pool layer to reduce dimension
      )
       self.classification_layer = nn.Sequential(
           # first we will used flatten layer to flatten tensor from convolutional layer
           nn.Flatten(),
           # hidden layer
           nn.Linear(64*7*7,128), # 64 by 7 by 7 is input dimension required for hidden layer, 128 number of neurons
           nn.ReLU(),
           nn.Dropout(p=0.4),
           nn.Linear(128,64),
           nn.ReLU(),
           nn.Dropout(p=0.4),
           nn.Linear(64,10), # output layer
           # did not pplied sigmoid because we are using cross entropy loss that internally uses sigmoid to convert classes to probabilites

           # tensor shape
           #first we use 2d image in conv layer of shape 1,28,28 where 28 by 28 pixel height and width and 1 represents gray scale image
            #  then this image is passed to 32 filters/feature maps ,so shape becomes 32,28,28 because same padding keeps shape of matrix/tensor same
             #  then pooling layer reduces dimension to half because 2 kernal size and 2 srtide,so shape becomes 32,14,14
              # then 2nd conv layer where 64 neurons passed made is 64,14,14 as shape,
               #which is then pass to wnd pool layer that half dimesnison to 64,7,7 '''
       )

     def forward(self,x):
        x= self.feature_extraction_layer(x)
        x=self.classification_layer(x)
        return x
```

```python
# setting learning rate and epochs
epochs = 50
learning_rate = 0.1
```

```python
# instansiate the model/make model object
model = Model(1) # providing no. of columns as argument to num_features parameter
model = model.to(device) # step 2 for training on GPU
```

```python
# loss function, its independent not used with model object
loss_function = nn.CrossEntropyLoss()
```

```python
# optimizer
optimizer = optim.SGD(model.parameters(),lr= learning_rate,weight_decay=1e-4) # 1e-4= 10**-4,applied l2 regularization
```

```python
# total batches
len(train_loader)
```

```python
# Training pipeline
# defining loop fpr each epoch
for epoch in range(epochs):
  total_epoch_loss = 0 # initialized, will help to check total loss per epoch
  # defining loop to load batches
  for batch_features , batch_labels in train_loader:
    # move data to GPU, step 3 for training model on gpu
    batch_features , batch_labels = batch_features.to(device) , batch_labels.to(device)
    # forward pass
    y_predicted = model(batch_features)
    # loss calculate
    loss = loss_function(y_predicted,batch_labels)
    # clear gradients
    optimizer.zero_grad()
    # back pass
    loss.backward()
    # parameters update
    optimizer.step()

    total_epoch_loss = total_epoch_loss + loss.item() # loss. item will convert single batch loss into scalar and add it to toal epoch loss, will run until loop end, loop will run for 150 batches
  average_loss = total_epoch_loss/len(train_loader)   # means average loss in each epoch per batch ,lentrain loder will give total no. of batches
  print(f'Epoch:{epoch + 1} , Loss: {average_loss}')
```

```python
# set model to evaluation mode
model.eval() # tells model that we use now this model for evaluation purpose
```

```python
# evaluation
total = 0
correct = 0

with torch.no_grad():
    for batch_features, batch_labels in test_loader:
        batch_features , batch_labels = batch_features.to(device) , batch_labels.to(device)

        # forward pass
        y_predicted = model(batch_features)

        # get predicted class indices
        _, predicted = torch.max(y_predicted, 1)
        # we dont need max value,we actually need index ,the position where it lies to extract class
        # so using _, means ignore first value, return 2nd vvalue baause normally it return both max and indices

        # total samples
        total = total + batch_labels.shape[0]

        # correct predictions
        correct = correct + (predicted == batch_labels).sum().item()

accuracy=(correct / total)*100
print(f'Accuracy={accuracy}%')
```

```python
# evaluation on training data

total = 0
correct = 0

with torch.no_grad():
    for batch_features, batch_labels in train_loader:
        batch_features , batch_labels = batch_features.to(device) , batch_labels.to(device)

        # forward pass
        y_predicted = model(batch_features)

        # get predicted class indices
        _, predicted = torch.max(y_predicted, 1)
        # we dont need max value,we actually need index ,the position where it lies to extract class
        # so using _, means ignore first value, return 2nd vvalue baause normally it return both max and indices

        # total samples
        total = total + batch_labels.shape[0]

        # correct predictions
        correct = correct + (predicted == batch_labels).sum().item()

accuracy=(correct / total)*100
print(f'Accuracy={accuracy}%')
```

```python
# since training accuracy is 93% and test accuracy is 83% approx,this 10% gap means ovefitting
# solution for optimization and reducing over fitting
# 1) adding more data, reduce chance of memorizing data,forced to lear patterns as the data become diverse,reduce biasness,not applicable in our case as we already used whole data
# 2) reducing complexity of neural network arcitechture,not applicable in our case as our arcitechture has only 2 layers
# 3) Regularization ,adding penalty to loss function,L1,L2,L2 is used mostly in deep learning,applicale in out case
# 4) Dropouts ,randomly shutting down few neuron of some layers,during training, applicable in our case
# 5) Data Augmentation , modifying images, tilt,flip,rotate etc, so variations, diversity increaces,better to use with cnn
# 6) Batch normalization ,to stabilize training,applicable
# 7) early stopping, stopping training early ,will not use
```

```python
# Dropout
# applied to hidden layers
# applied after relu activation fucntion
# randomly turns p% neurons in hidden layer during each forward pass,we set p value ,for example p=0.5 means turn off 50 % neurons
# this has regularization effect
# during evaluation,dropout is not used,
```

```python
# Batch normalization
# improves training stability and speed up training
# parameters keeps updating during each forward pass,we normalize inputs of each layer   between certain range of values before passing them between hidden layers, during training
# typically applies to hidden layers
# solve the problem of  internal covariate shift,means next layer keeps seeing different distributions,cause training becomes slow,network becomes unstable
# batch normalization takes activation and does measn= 0 and variance= 1,now data is centered scaled stable
# applied after hidden layer;s linear layer, ,before activation function
# includes learnable parameters like scaling,shifting
# not used during EVALUATION
# normalize activation means normalize features because 1 neuron that get weights and bias represent 1 feature/column
# distribution means centermean of datata and its spread variance
```

```python
# Regularization
# we add sum penalty term in loss function, so then we are focused not only minimizing loss but combination of loss and penalty term
# results in not getting high values of weights, this helps with over fitting
# applied to model wights, not on biases
# example in l2 regularization the penalty term is sum of, square of all the weights, multplied with regularization constant lambda,addedto orignal loss function
# lambda constant tells how strong regularization you want
# we can simply add loss term to gradients during back pass, this is called weights decay,
# it basically penalizes large weights,so big values of weight are distributed to other weights thus ove dependece on certain weights decreaces
# no effect on bias,only applied during training
# we are going to apply l2 regularization furing back pass to our gradients,simply weights decay
```
